# 02 – Brecha entre regímenes (Acto 1)

Este cuaderno prepara cálculos base de participación e índices para comparar régimen especial y definitivo en el PIB.

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

data_dir = Path('../data/processed')
df_corrientes = pd.read_csv(data_dir / 'bccr-pib-corrientes-limpio.csv')
df_encadenados = pd.read_csv(data_dir / 'bccr-pib-encadenados-limpio.csv')
```

## Cálculo de brechas

```python
df_corrientes['pct_especial'] = df_corrientes['regimen_especial'] / df_corrientes['pib_total'] * 100
df_corrientes['pct_definitivo'] = df_corrientes['regimen_definitivo'] / df_corrientes['pib_total'] * 100
df_corrientes = df_corrientes.query('año >= 2000 and año <= 2026').copy()

df_encadenados['pct_especial'] = df_encadenados['regimen_especial'] / df_encadenados['pib_total'] * 100
df_encadenados['pct_definitivo'] = df_encadenados['regimen_definitivo'] / df_encadenados['pib_total'] * 100
df_encadenados = df_encadenados.query('año >= 2000 and año <= 2026').copy()
```

## Gráfico 1: participación del régimen especial en el PIB (encadenados)

```python
plt.figure(figsize=(10, 5))
plt.plot(df_encadenados['año'], df_encadenados['pct_especial'], marker='o', linewidth=1.8)
plt.title('Participación del régimen especial en el PIB (encadenados), 2000-2026')
plt.xlabel('Año')
plt.ylabel('Porcentaje del PIB (%)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Gráfico 2: índice de crecimiento base 2000 (encadenados)

```python
base_2000 = df_encadenados.loc[df_encadenados['año'] == 2000, ['pib_total', 'regimen_definitivo', 'regimen_especial']].iloc[0]

idx_pib_total = df_encadenados['pib_total'] / base_2000['pib_total'] * 100
idx_regimen_definitivo = df_encadenados['regimen_definitivo'] / base_2000['regimen_definitivo'] * 100
idx_regimen_especial = df_encadenados['regimen_especial'] / base_2000['regimen_especial'] * 100

plt.figure(figsize=(10, 5))
plt.plot(df_encadenados['año'], idx_pib_total, label='PIB total', linewidth=1.8)
plt.plot(df_encadenados['año'], idx_regimen_definitivo, label='Régimen definitivo', linewidth=1.8)
plt.plot(df_encadenados['año'], idx_regimen_especial, label='Régimen especial', linewidth=1.8)
plt.title('Índice de crecimiento base 2000 (encadenados), 2000-2026')
plt.xlabel('Año')
plt.ylabel('Índice (base 100 = año 2000)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
